

### 2.Visualization
The visualization is conducted separately for the three sets of data we collected.
We used three types of visualizations to observe and compare the data.


1. **Histograms**:  
   - For each participant (Participant 1, Participant 2, Participant 3), we plotted histograms where the x-axis represents the amount of the red envelopes and the y-axis shows the corresponding frequency, using 0.1 intervals. This allows us to visually observe the frequency distribution of amounts received by different participants and identify any frequently occurring amounts or outliers.  

2. **Box Plots**:  
   - We also created a box plot to showcase the distribution of amounts for the three participants using different colors. The box plot effectively displays each participant's median, quartiles, and potential outliers, helping us analyze the differences in amounts received among participants.  

3. **Scatter Plots**:  
   - Finally, to illustrate individual data points, we plotted scatter plots for each participant. The x-axis represents the trial number, while the y-axis shows the specific red envelope amounts. This visualization helps us understand the variations in each participant's performance across different trials, revealing patterns in the amounts over time.  


```python
import pandas as pd  
import matplotlib.pyplot as plt  
import numpy as np  

# Read the CSV file  
df = pd.read_csv('data1.csv')  # Updated file name  

# View original column names  
print("Original column names:", df.columns.tolist())  

# Rename columns for easier access  
df.columns = ['Participant1', 'Participant2', 'Participant3']  

# Create figures  
plt.figure(figsize=(15, 15))  

# Histograms with 0.1 bin width  
plt.subplot(3, 3, 1)  
plt.hist(df['Participant1'], bins=np.arange(0, 5.1, 0.1), edgecolor='black')  # 0.1 intervals  
plt.title('Histogram of Participant 1')  
plt.xlabel('Value')  
plt.ylabel('Frequency')  

plt.subplot(3, 3, 2)  
plt.hist(df['Participant2'], bins=np.arange(0, 5.1, 0.1), edgecolor='black')  # 0.1 intervals  
plt.title('Histogram of Participant 2')  
plt.xlabel('Value')  
plt.ylabel('Frequency')  

plt.subplot(3, 3, 3)  
plt.hist(df['Participant3'], bins=np.arange(0, 5.1, 0.1), edgecolor='black')  # 0.1 intervals  
plt.title('Histogram of Participant 3')  
plt.xlabel('Value')  
plt.ylabel('Frequency')  

# Box plots on the same plot with different colors  
plt.subplot(3, 3, 4)  
box_data = [df['Participant1'], df['Participant2'], df['Participant3']]  
box_colors = ['lightblue', 'lightgreen', 'salmon']  
box_positions = [1, 2, 3]  

# Create box plots with patch_artist to allow color fill  
box_parts = plt.boxplot(box_data, positions=box_positions, patch_artist=True, widths=0.6)  

# Set colors for each box  
for i, patch in enumerate(box_parts['boxes']):  
    patch.set_facecolor(box_colors[i])  # Set the face color  
    patch.set_edgecolor('black')  # Set the edge color  

plt.title('Box Plot Comparison of Participants')  
plt.ylabel('Value')  
plt.xticks(box_positions, ['Participant 1', 'Participant 2', 'Participant 3'])  

# Scatter plots  
plt.subplot(3, 3, 7)  
plt.scatter(range(len(df['Participant1'])), df['Participant1'])  
plt.title('Scatter Plot of Participant 1')  
plt.xlabel('Trial Number')  
plt.ylabel('Value')  

plt.subplot(3, 3, 8)  
plt.scatter(range(len(df['Participant2'])), df['Participant2'])  
plt.title('Scatter Plot of Participant 2')  
plt.xlabel('Trial Number')  
plt.ylabel('Value')  

plt.subplot(3, 3, 9)  
plt.scatter(range(len(df['Participant3'])), df['Participant3'])  
plt.title('Scatter Plot of Participant 3')  
plt.xlabel('Trial Number')  
plt.ylabel('Value')  

# Adjust subplot spacing  
plt.tight_layout()  

# Show the plots  
plt.show()
```
